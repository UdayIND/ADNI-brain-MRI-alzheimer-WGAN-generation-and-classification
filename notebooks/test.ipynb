```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sentences = ["This is the first sentence.", "This is the second sentence.", "This is the third sentence.", "This is the fourth sentence.", "This is the fifth sentence."]
social_subjects = pd.Series(np.array(sentences * 1000))
```

```python
social_subjects
```

```python
fig, ax = plt.subplots()
ax.bar(social_subjects.value_counts().index,
        social_subjects.value_counts())
plt.xticks(rotation='vertical')
ax.set_xticklabels(social_subjects.value_counts().index, rotation=0)
plt.title('Home accomodation distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import textwrap

# Dati di esempio
labels = ['House', 'Condo/Co-op (owned)', 'Apartment (rented)', 'Retirement Community', 'Assisted Living Living', 'Assisted Living', 'Skille Nursing Facility','Assisted Living' ]
values = [10, 20, 15, 5, 5, 5, 5, 5]

# Creazione del barplot
fig, ax = plt.subplots()
ax.bar(range(len(values)), values, width=0.8)

# Personalizzazione delle etichette dell'asse x
wrapped_labels = [textwrap.fill(label, width=15) for label in labels]
wrapped_labels = [label.replace(' ', '\n') for label in wrapped_labels]
wrapped_labels = [label.replace('/', '\n') for label in wrapped_labels]
plt.xticks(range(len(labels)), wrapped_labels, rotation=45)

# Mostra il grafico
plt.show()
```
